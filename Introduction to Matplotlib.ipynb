## Import Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [1, 2, 3]
y = [2, 3, 4]

plt.plot(x, y, color = '#01acab')
plt.title('First graph in Matplotlib!', fontdict={'fontname': 'comic sans MS', 'fontsize': 20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

#Keywork Argument Notation
plt.plot(x, y, label ='2x', color = 'red', linewidth = '2', marker = '.', markersize = 10, markeredgecolor = 'blue' ,linestyle ='--')
plt.title('Our First graph!', fontdict={'fontname': 'comic sans MS', 'fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

#Determining the X-ticks and the Y-ticks
#X and Y Tickmarks (scale of your graph)
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])

#add a legend
plt.legend()
plt.show()
```

```python


x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]



#to resize your graph (dpi specifies pixel per inch, when saving, you should probably use 300 if possible)
plt.figure(figsize = (5, 3), dpi = 150)

#Line 1

#Using Shorthand Notation
#fmt ='[color][marker][Linestyle]'
plt.plot(x, y, 'r.--', label ='2x')

#Add a title (Specify font parameters with fontdict)
plt.title('Our First graph!', fontdict={'fontname': 'comic sans MS', 'fontsize':20})

## X and Y label
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')



#Line 2
## Creating a second line using Numpy
#Select interval we want to plot points at
x2 = np.arange(0, 4.5, 0.5)
#plt.plot(x2, x2**2, 'b', label = 'x2')


#Here we want to plot the first 6 values and plot normal and plot the rest with dashed line
#Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'b', label = 'x2')

#Plot part of the graph as dashed line
plt.plot(x2[5:], x2[5:]**2, 'b--')
#print(x2)

#add a legend
plt.legend()


#Saving figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi = 300)

#show plot
plt.show()
```

## Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)

#to set patterns in each bar individually
#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

#You can also set the patterns using shortcut
patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))


plt.title('My First Bar Chart')
plt.xlabel('Labels')
plt.ylabel('values')

#plt.legend()
plt.figure(figsize = (5, 3))
plt.show()
```
